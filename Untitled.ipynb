```python
# import neccessary lib
import cv2
from PIL import Image, ImageOps
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
# ulity function
def img_data(imgPath,disp = True):
    
    orig_img = Image.open(imgPath)
    
    img_size_kb = os.stat(imgPath).st_size/1024
    
    ori_pixels = np.array(orig_img.getdata()).reshape(*orig_img.size, -1)
    
    img_dim = ori_pixels.shape 
    
    if disp:
        plt.imshow(orig_img)
        plt.show()
    
    data_dict = {}
    data_dict['img_size_kb'] = img_size_kb
    data_dict['img_dim'] = img_dim
    
    return data_dict
```

```python
# load origin image and show information
imgPath = "SampleJPGImage.jpg"
img = cv2.imread(imgPath)

data_dict_ori = img_data(imgPath)

print('Original Image Data')
print('Original Image size (kB)',data_dict_ori['img_size_kb'])
print('Original Image Shape',data_dict_ori['img_dim'])
```

```python
# split the image into 3 chennels
r, g, b = cv2.split(img)
```

```python
# change scale to show image in notebook
#r, g, b = r/255, g/255, b/255
```

```python
# do the pca for 3 channels
pca_comp = 200

pca_r = PCA(n_components = pca_comp)
reduced_r = pca_r.fit_transform(r)

pca_g = PCA(n_components = pca_comp)
reduced_g = pca_g.fit_transform(g)

pca_b = PCA(n_components = pca_comp)
reduced_b = pca_b.fit_transform(b)
```

```python
# reconstruct 3 channels and merge into an image

reconstructed_r = pca_r.inverse_transform(reduced_r)
reconstructed_g = pca_g.inverse_transform(reduced_g)
reconstructed_b = pca_b.inverse_transform(reduced_b)

img_reconstructed = (cv2.merge((reconstructed_r,reconstructed_g, reconstructed_b)))
```

```python
# write the result to file and how it is compressed
imgPath_comp = "SampleJPG_Compressed.jpg"
cv2.imwrite(imgPath_comp, img_reconstructed)
```

```python
# show data of the compressed image
data_dict_comp = img_data(imgPath_comp)

print('Compressed Image Data')
print('Compressed Image size (kB)',data_dict_comp['img_size_kb'])
print('Compressed Image Shape',data_dict_comp['img_dim'])
```
